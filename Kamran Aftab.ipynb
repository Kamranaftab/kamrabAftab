```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt  # Correct import statement for plotting

# Load the CSV file into DataFrame
df = pd.read_csv("data.csv")

# Filter the data for the specific year (2014)
year_data = df[df['year'] == 2014]

# Plot the 'value' over 'month' for the year 2014
plt.figure(figsize=(10, 6))  # Create a figure
plt.plot(year_data['month'], year_data['value'], marker='o', label='Value in 2014', color='blue')  # Plot the line graph
plt.xlabel('Month')  # Label for the X-axis
plt.ylabel('Value (MMCF)')  # Label for the Y-axis
plt.title('Monthly Natural Gas Deliveries in 2014')  # Title of the plot
plt.legend()  # Display the legend
plt.grid(True)  # Show gridlines
plt.show()  # Display the plot
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Load the CSV file into DataFrame
df = pd.read_csv("data.csv")

# Strip spaces from column names, if any
df.columns = df.columns.str.strip()

# Filter data for the specific year (e.g., 2014)
year_data = df[df['year'] == 2014]  # Use the correct column name ('year' instead of 'Year')

# Group the data by 'area-name' and sum the 'value' for each area
area_values = year_data.groupby('area-name')['value'].sum()

# Create a pie chart based on the total value for each area
plt.figure(figsize=(8, 8))
plt.pie(area_values, labels=area_values.index, autopct='%1.1f%%', startangle=140, colors=plt.cm.Paired.colors)
plt.title('Distribution of Natural Gas Deliveries by Area in 2014')
plt.show()
```
